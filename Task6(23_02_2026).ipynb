```python
def parking_system(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Full (Alert!)")
    else:
        print("Parking Status: Available")

# Sample Input
parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
def delivery_time(distance, traffic, weather):
    time = distance * 5   # base time (5 mins per km)

    # Traffic adjustment
    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 10

    # Weather adjustment
    if weather == "Rainy":
        time += 10
    elif weather == "Storm":
        time += 20

    print("Estimated Delivery Time:", time, "minutes")

# Sample Input
delivery_time(8, "High", "Rainy")
```

```python
def theatre_occupancy(total_seats, booked_seats):
    booked = len(booked_seats)
    occupancy = (booked / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Available")

# Sample Input
theatre_occupancy(200, [1]*150)
```

```python
def server_load(cpu_readings):
    total = 0

    for value in cpu_readings:
        total += value

    avg = total / len(cpu_readings)

    print("Average CPU Load:", int(avg), "%")

    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

# Sample Input
server_load([45, 60, 70, 85, 90])
```

```python
def resource_monitor(resources):
    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    print("Overused Resources:", ", ".join(overused))

    if overused:
        print("Energy Alert: Yes")
    else:
        print("Energy Alert: No")

# Sample Input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(usage)
```

```python
def event_registration(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlist = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)

# Sample Input
event_registration(100, 105)
```
